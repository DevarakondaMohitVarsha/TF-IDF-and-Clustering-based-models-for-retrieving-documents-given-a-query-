## K-means with Cosine Similarity
* assuming 10 clusters and 100 iterations
* Full manual implementation of k-means cosine similarity

```python
import math
import numpy

def csim(x,y):
    return (    numpy.dot(x,y)/(numpy.linalg.norm(x)*numpy.linalg.norm(y))   )


def kmeans_cosine(k,corp,iterations):
    seeds=[]
    for x in range(0,k):
        seeds.append(x*(math.floor(1014/k)))
    #print(seeds)
    final_labels = [-1]*len(corp)
    centroids = numpy.ndarray( (k,len(corp[0])) )
    #initializing centroids
    for z1 in range(0,k):
        centroids[z1]=corp[seeds[z1]]
    
    
    for z2 in range(0,iterations):
        temp_clist = [[] for i in range(k)]
        cent_sum = numpy.zeros( (k,len(corp[0])) )
        for z3 in range(0,len(corp)):
            tmax=0.0
            tind=0
            cs=0.0
            for z4 in range(0,k):
                cs=csim(centroids[z4],corp[z3])
                if(cs>=tmax):
                    tmax = cs
                    tind = z4
            final_labels[z3]=tind
            temp_clist[final_labels[z3]].append(z3)
            cent_sum[final_labels[z3]]+=corp[z3]
        for z5 in range(0,k):
            #print(len(temp_clist[z5]))
            if(len(temp_clist[z5])!=0): 
                cent_sum[z5]=cent_sum[z5]/len(temp_clist[z5])
            else:
                cent_sum[z5]=centroids[z5]
        centroids = cent_sum
        return_kcos = {}
        return_kcos['centroids']=centroids
        return_kcos['k_value']=k
        return_kcos['labels']=final_labels
        
    return return_kcos

clusters_kcosine = kmeans_cosine(10,vect_corp_tfidf,100)
print(clusters_kcosine['labels'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity as cos_sim
from sklearn.cluster import AgglomerativeClustering
import numpy as np


def agglomerative_single(k,vctfidf,link):
    Agg_sin = AgglomerativeClustering(n_clusters=k,affinity = 'euclidean',linkage=link)
    best_asin = vctfidf
    Agg_sin.fit_predict(best_asin)
    clusters_asin = Agg_sin.labels_.tolist()
    #centroids_best_keu = kmeans_model.cluster_centers_
    final_cosine = cos_sim(vctfidf) # calculates cosine similarities and stores them as 1014x1014

    best_asin_list = [[] for i in range(k)]
    intra_sum_asin = 0.0
    intra_avg_asin = 0.0
    intra_comb_asin = 0.0

    for x31 in range(0,len(clusters_asin)):
        best_asin_list[clusters_asin[x31]].append(x31)

    for x33 in range(0,len(best_asin_list)):
        intra_sum_asin = 0.0
        for x34 in range(0,len(best_asin_list[x33])-1):
            for x35 in range(x34+1,len(best_asin_list[x33])):
                intra_sum_asin = intra_sum_asin + final_cosine[best_asin_list[x33][x34]][best_asin_list[x33][x35]]
        intra_comb_asin = (len(best_asin_list[x33])*(len(best_asin_list[x33])-1))/2
        if(len(best_asin_list[x33])==1):
            intra_comb_asin=1
        intra_avg_asin += intra_sum_asin/intra_comb_asin
        #print(intra_sum_asin/intra_comb_asin)

    #print("\n")
    #print(intra_avg_asin/k)

    # using centroids to calculate inter cluster similarities
    #cent_eu = kmeans_model.cluster_centers_
    
    
    centroids_asin = np.ndarray((k,len(best_asin[0])))
    
    
    for c1 in range(0,len(best_asin_list)):
        cent_temp_asin = np.zeros((1,len(best_asin[0])))
        for c2 in range(0,len(best_asin_list[c1])):
            cent_temp_asin=cent_temp_asin+best_asin[c2]
        cent_temp_asin/len(best_asin_list[c1])
        centroids_asin[c1]=cent_temp_asin
            
    
    #print(centroids_asin)
    cos_cent_asin = cos_sim(centroids_asin)
    
    inter_sum_asin = 0.0
    inter_avg_asin= 0.0
    inter_comb_asin = ((k)*(k-1))/2
    for x36 in range(0,k-1):
        inter_sum_asin=0
        for x37 in range(x36+1,k):
            inter_sum_asin = inter_sum_asin+cos_cent_asin[x36][x37]
        #print(inter_sum_asin/inter_comb_asin)
    inter_avg_asin += inter_sum_asin/inter_comb_asin
    #print("\n")
    #print(inter_avg_asin)
    #print((intra_avg_asin-inter_avg_asin),' ',(abs(intra_avg_asin-inter_avg_asin)/max(intra_avg_asin,inter_avg_asin)),' ',(intra_avg_asin/inter_avg_asin))
    #return (intra_avg_asin-inter_avg_asin)
    
    #print(abs(intra_avg_asin-inter_avg_asin)/max(intra_avg_asin,inter_avg_asin))
    #return (abs(intra_avg_asin-inter_avg_asin)/max(intra_avg_asin,inter_avg_asin))
    score = intra_avg_asin-inter_avg_asin
    return_agg = {}
    return_agg['intra']=intra_avg_asin
    return_agg['inter']=inter_avg_asin
    return_agg['score']=score
    return_agg['centroids']=centroids_asin
    return_agg['k_value']=k
    return_agg['labels']=clusters_asin
    return return_agg
```

* Best k calculator implementation

```python
def inter_intra_scores(start,stop,corp,algo):
    tlist =[]
    max_keu =0.0
    for x38 in range(start,stop):
        if(algo=='kmeans_eu_score'):
            keu_out = kmeans_euclidean(x38,vect_corp_tfidf)
        elif(algo == 'kmeans_cosine_score'):
            keu_out = kmeans_cosine_similarity(x38,vect_corp_tfidf)
        elif(algo== 'single'):
            keu_out = agglomerative_single(x38,vect_corp_tfidf,link=algo)
        elif(algo=='complete'):
            keu_out = agglomerative_single(x38,vect_corp_tfidf,link=algo)
        elif(algo=='average'):
            keu_out = agglomerative_single(x38,vect_corp_tfidf,link=algo)
        elif(algo=='ward'):
            keu_out = agglomerative_single(x38,vect_corp_tfidf,link=algo)
        
        tlist.append((keu_out['intra'],keu_out['inter'],keu_out['score']))
        if( keu_out['score'] >= max_keu) :
            max_keu = keu_out['score']
    #print(max_keu)
    #print("\n")
    #print(tlist)
    max_score_index = 0
    for x39 in range(0,len(tlist)):
        if(max_keu == tlist[x39][2]):
            max_score_index = x39+start
            
    return_scores ={}
    return_scores['score']=tlist
    return_scores['max_score']=max_keu
    return_scores['max_score_index']=max_score_index
    
    return return_scores
```

```python
def graph_plotter(inp,offset):
    intra=[]
    for x in range(0,len(inp)):
        intra.append(inp[x][0])
    inter=[]
    for y in range(0,len(inp)):
        inter.append(inp[y][1])
    k=[]
    for z in range(0,len(inp)):
        k.append(z+offset)
    ret_graph=[]
    ret_graph.append(intra)
    ret_graph.append(inter)
    ret_graph.append(k)
    return ret_graph
```

* prec_rec_final function computes precision and recall for any of the 6 clustering algorithms given as parameter along with the k value and, returns the results in the form of list of lists

```python
def pre_rec_final(method,k,vect_query_tfidf,vect_corp_tfidf,task2_ov_p,task2_ov_r):
    if(method == 'kmeans_eu'):
        queries_kmeans_eu = vect_query_tfidf
        best_keu_result = kmeans_euclidean(k,vect_corp_tfidf) 
    if(method == 'kmeans_cosine'):
        queries_kmeans_eu = vect_query_tfidf
        best_keu_result = kmeans_cosine_similarity(k,vect_corp_tfidf) 
    if(method == 'single'):   
        queries_kmeans_eu = vect_query_tfidf
        best_keu_result = agglomerative_single(k,vect_corp_tfidf,method)
    if(method == 'complete'):   
        queries_kmeans_eu = vect_query_tfidf
        best_keu_result = agglomerative_single(k,vect_corp_tfidf,method)
    if(method == 'average'):   
        queries_kmeans_eu = vect_query_tfidf
        best_keu_result = agglomerative_single(k,vect_corp_tfidf,method)
    if(method == 'ward'):   
        queries_kmeans_eu = vect_query_tfidf
        best_keu_result = agglomerative_single(k,vect_corp_tfidf,method)

        
    query_keu_scores = numpy.zeros( ( len(vect_query_tfidf) , len(best_keu_result['centroids']) ) )
    for x501 in range(0,len(vect_query_tfidf)):
        for x502 in range(0,len(best_keu_result['centroids'])):
            query_keu_scores[x501][x502] = csim(vect_query_tfidf[x501],best_keu_result['centroids'][x502])
    final_keu_scores = []
    for x503 in range(0,len(query_keu_scores)):
        final_keu_scores.append(numpy.where(query_keu_scores[x503]==numpy.amax(query_keu_scores[x503])))
            
    final_retrieved = [[] for i in range(0,10)]
    for x504 in range(0,10):
        for x505 in range(0,1014):
            if(final_keu_scores[x504][0][0]==best_keu_result['labels'][x505]):
                final_retrieved[x504].append(x505)
                    
    final_precision = []
    final_recall = []
    for x510 in range(0,10):
        final_precision.append(len((set(docs_rel_id[x510]).intersection(set(final_retrieved[x510]))))/len(final_retrieved[x510]))
        final_recall.append(len((set(docs_rel_id[x510]).intersection(set(final_retrieved[x510]))))/relevant_documents[x510])  
            
    pre_rec = []
    pre_rec.append(task2_ov_p)
    pre_rec.append(task2_ov_r)
    pre_rec.append(final_precision)
    pre_rec.append(final_recall)
    
    return pre_rec
```
